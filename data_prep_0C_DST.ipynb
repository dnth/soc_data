```python
from pandas import read_excel
from datetime import datetime
```

```python
# load data
def parse(x):
	return datetime.strptime(x, '%Y %m %d %H')
```

```python
dataset = read_excel('../data/raw_data/DST/SP2_0C_DST/02_24_2016_SP20-2_0C_DST_80SOC.xls',  'Channel_1-006')
```

```python
dataset
```

```python
dataset = dataset.drop(['Date_Time','Test_Time(s)','Data_Point', 'Step_Index', 'Step_Time(s)', 'Cycle_Index', 'Charge_Capacity(Ah)', 'Discharge_Capacity(Ah)', 'Charge_Energy(Wh)', 'Discharge_Energy(Wh)', 'dV/dt(V/s)', 'Internal_Resistance(Ohm)', 'Is_FC_Data', 'AC_Impedance(Ohm)', 'ACI_Phase_Angle(Deg)'], axis=1)
```

```python
dataset
```

```python
# add SOC column
dataset['SOC'] = 0.8
```

```python
dataset
```

```python
# SOC calculation
for i in range(1, len(dataset)):
    dataset.loc[i, 'SOC'] = dataset.loc[i-1, 'SOC'] + dataset.loc[i-1, 'Current(A)']/7200
```

```python
dataset
```

```python
% matplotlib inline
import seaborn as sns
sns.set()
dataset.plot(figsize=(20,10))
```

```python
# remove the first 2000 rows
dataset = dataset[2000:]
```
